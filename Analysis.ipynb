```python
import pandas as pd 

df_orders=pd.read_csv('olist_orders_dataset.csv')
df_items=pd.read_csv('olist_order_items_dataset.csv')
df_reviews=pd.read_csv('olist_order_reviews_dataset.csv')

df_merged=pd.merge(df_items,df_orders,on='order_id',how='inner')

df_final=pd.merge(df_merged,df_reviews,on='order_id',how='inner')

print(f":{df_final.shape}")
df_final.head()
```

```python
print(df_final.columns.tolist())
```

```python
print(df_final.isnull().sum())
```

```python
df_clean=df_final.dropna(subset=['review_comment_message'])
print(f":{df_clean.shape}")
                         
```

```python
print(df_clean['review_score'].value_counts())
```

```python
import matplotlib.pyplot as plt

df_clean['review_score'].value_counts().sort_index().plot(kind='bar')
plt.title("Review Score Distirubtion")
plt.xlabel("Score")
plt.ylabel("Count")
plt.show()

```
